```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import re
from tqdm.notebook import tqdm
import torch.nn.functional as F
import random
import kagglehub
from pathlib import Path
from torch.nn.utils import spectral_norm

# 시드 고정
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    
set_seed()

# 장치 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# UTKFace 데이터셋 클래스 정의
class UTKFaceDataset(Dataset):
    def __init__(self, root_dir, transform=None, min_age=1, max_age=100):
        self.root_dir = Path("/kaggle/input/utkface-new/crop_part1")
        self.transform = transform
        self.min_age = min_age
        self.max_age = max_age
        
        # 파일 목록 가져오기
        self.image_files = []
        for filename in os.listdir(root_dir):
            if filename.endswith('.jpg') or filename.endswith('.png'):
                # UTKFace 파일명 형식: [age]_[gender]_[race]_[date&time].jpg
                try:
                    age = int(filename.split('_')[0])
                    if min_age <= age <= max_age:
                        self.image_files.append(filename)
                except:
                    continue
        
        print(f"Total valid images: {len(self.image_files)}")
    
    def __len__(self):
        return len(self.image_files)
    
    def __getitem__(self, idx):
        img_name = os.path.join(self.root_dir, self.image_files[idx])
        image = Image.open(img_name).convert('RGB')
        
        # 나이 추출
        age = int(self.image_files[idx].split('_')[0])
        # 정규화된 나이 (0-1 사이)
        age_normalized = (age - self.min_age) / (self.max_age - self.min_age)
        
        if self.transform:
            image = self.transform(image)
            
        return image, torch.tensor([age_normalized], dtype=torch.float)

# 데이터 전처리 정의
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# 데이터셋 로드 (Kaggle UTKFace 경로에 맞게 수정)
dataset = UTKFaceDataset(root_dir="/kaggle/input/utkface-new/crop_part1", transform=transform)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True, num_workers=2)

# Residual Block 정의
class ResBlock(nn.Module):
    def __init__(self, channels):
        super(ResBlock, self).__init__()
        self.block = nn.Sequential(
            nn.Conv2d(channels, channels, 3, 1, 1, bias=False),
            nn.BatchNorm2d(channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(channels, channels, 3, 1, 1, bias=False),
            nn.BatchNorm2d(channels)
        )
    
    def forward(self, x):
        return x + self.block(x)

# 생성자 네트워크 정의
class Generator(nn.Module):
    def __init__(self, latent_dim=100, age_dim=1):
        super(Generator, self).__init__()
        self.latent_dim = latent_dim
        
        # 입력층: 노이즈 + 나이 정보
        self.fc = nn.Sequential(
            nn.Linear(latent_dim + age_dim, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(True),

            nn.Linear(1024, 512 * 4 * 4),
            nn.BatchNorm1d(512 * 4 * 4),
            nn.ReLU(True)
        )
        
        self.main = nn.Sequential(
            # 4x4
            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),  # 8x8
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            ResBlock(256),

            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),  # 16x16
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            ResBlock(128),

            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),   # 32x32
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),     # 64x64
            nn.Tanh()
        )

    def forward(self, z, age):
        x = torch.cat([z, age], dim=1)
        x = self.fc(x)
        x = x.view(-1, 512, 4, 4)
        return self.main(x)

# Minibatch Discrimination Layer 정의
class MinibatchDiscrimination(nn.Module):
    def __init__(self, in_features, out_features, kernel_dim=50):
        super().__init__()
        self.T = nn.Parameter(torch.randn(in_features, out_features, kernel_dim))

    def forward(self, x):
        M = x.matmul(self.T.view(x.size(1), -1))
        M = M.view(x.size(0), -1, self.T.size(1))  # [B, out_features, kernel_dim]

        out = []
        for i in range(M.size(0)):
            diff = M[i].unsqueeze(0) - M  # [B, out_features, kernel_dim]
            diff = torch.abs(diff).sum(2)  # L1 distance across kernel_dim
            out.append(torch.exp(-diff).sum(0))  # sum across batch dim

        out = torch.stack(out)  # [B, out_features]
        return torch.cat([x, out], dim=1)
        
# 판별자 네트워크 정의
class Discriminator(nn.Module):
    def __init__(self, age_dim=1):
        super(Discriminator, self).__init__()
        
        # 이미지 처리 부분
        self.conv = nn.Sequential(
            spectral_norm(nn.Conv2d(3, 64, 4, 2, 1, bias=False)),  # 32x32
            nn.LeakyReLU(0.2, inplace=True),

            spectral_norm(nn.Conv2d(64, 128, 4, 2, 1, bias=False)),  # 16x16
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            spectral_norm(nn.Conv2d(128, 256, 4, 2, 1, bias=False)),  # 8x8
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),

            spectral_norm(nn.Conv2d(256, 512, 4, 2, 1, bias=False)),  # 4x4
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),
        )
        
        # 나이 정보 처리
        self.age_embedding = spectral_norm(nn.Linear(age_dim, 4 * 4))

        # 최종 판별 전에 Minibatch Discrimination
        self.mbd = MinibatchDiscrimination(in_features=512 * 4 * 4 + 4 * 4, out_features=50)

        # 최종 판별
        self.output = nn.Sequential(
            spectral_norm(nn.Linear(512 * 4 * 4 + 4 * 4 + 50, 1)),
            nn.Sigmoid()
        )
    
    def forward(self, x, age):
        batch_size = x.size(0)
        conv_out = self.conv(x)                     # [B, 512, 4, 4]
        conv_flat = conv_out.view(batch_size, -1)   # [B, 512*4*4]

        age_embedded = self.age_embedding(age)      # [B, 16]
        age_flat = age_embedded.view(batch_size, -1)

        features = torch.cat([conv_flat, age_flat], dim=1)  # [B, 512*16 + 16]
        features = self.mbd(features)                       # Add minibatch stats

        output = self.output(features)
        return output

    def extract_features(self, x, age):
        batch_size = x.size(0)
        conv_out = self.conv(x)                     # [B, 512, 4, 4]
        conv_flat = conv_out.view(batch_size, -1)   # [B, 512*4*4]

        age_embedded = self.age_embedding(age)      # [B, 16]
        age_flat = age_embedded.view(batch_size, -1)

        features = torch.cat([conv_flat, age_flat], dim=1)  # [B, 512*16 + 16]
        return features  # ⚠️ MBD를 적용하기 전의 특징 벡터

# 모델 초기화
latent_dim = 100
generator = Generator(latent_dim).to(device)
discriminator = Discriminator().to(device)

# 가중치 초기화 함수
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

# 가중치 초기화 적용
generator.apply(weights_init)
# discriminator.apply(weights_init) # SpectralNorm 적용으로 인한 비활성화

# 손실 함수와 옵티마이저 설정
criterion = nn.BCELoss()
optimizer_G = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))

# 학습 함수
def train_gan(epochs=200):
    # 학습 진행 기록용
    G_losses = []
    D_losses = []
    fixed_noise = torch.randn(64, latent_dim, device=device)
    
    # 샘플 나이 범주 생성 (0.2, 0.4, 0.6, 0.8에 해당하는 나이)
    sample_ages = torch.tensor([[0.2], [0.4], [0.6], [0.8]], device=device)

    # Historical Averaging: 평균 가중치 초기화
    avg_gen_weights = [p.detach().clone() for p in generator.parameters()]
    avg_decay = 0.9
    
    # 훈련 시작
    print("Starting Training Loop...")
    for epoch in range(epochs):
        for i, (real_images, ages) in enumerate(tqdm(dataloader)):
            batch_size = real_images.size(0)
            real_images = real_images.to(device)
            ages = ages.to(device)
            
            # 실제 이미지 레이블 (1)
            real_label = torch.full((batch_size, 1), 0.9, device=device)
            # 가짜 이미지 레이블 (0)
            fake_label = torch.rand(batch_size, 1, device=device) * 0.1
            
            # 판별자 학습
            optimizer_D.zero_grad()
            
            # 실제 이미지에 대한 손실
            output_real = discriminator(real_images, ages)
            loss_real = criterion(output_real, real_label)
            
            # 가짜 이미지에 대한 손실
            noise = torch.randn(batch_size, latent_dim, device=device)
            fake_images = generator(noise, ages)
            output_fake = discriminator(fake_images.detach(), ages)
            loss_fake = criterion(output_fake, fake_label)
            
            # 판별자 전체 손실
            loss_D = loss_real + loss_fake
            loss_D.backward()
            optimizer_D.step()
            
            # 생성자 학습
            optimizer_G.zero_grad()
            output = discriminator(fake_images, ages)
            gen_target_label = torch.full((batch_size, 1), 0.9, device=device)  # same as smoothed real label

            with torch.no_grad():
                real_feat = discriminator.extract_features(real_images, ages)
                fake_feat = discriminator.extract_features(fake_images, ages)
            feature_matching_loss = F.mse_loss(fake_feat, real_feat)

            historical_loss = 0
            for p, avg_p in zip(generator.parameters(), avg_gen_weights):
                historical_loss += torch.mean((p - avg_p.detach())**2)

            loss_G = criterion(output, gen_target_label)
            loss_G += 0.1 * feature_matching_loss  # Feature Matching 보조 손실
            loss_G += 0.1 * historical_loss # Historical Averaging 보조 손실

            loss_G.backward()
            optimizer_G.step()

            # 평균 가중치 업데이트
            with torch.no_grad():
                for p, avg_p in zip(generator.parameters(), avg_gen_weights):
                    avg_p.mul_(avg_decay).add_((1 - avg_decay) * p.detach())
            
            # 학습 상태 저장
            if i % 50 == 0:
                print(f"[{epoch}/{epochs}][{i}/{len(dataloader)}] Loss_D: {loss_D.item():.4f} Loss_G: {loss_G.item():.4f}")
                G_losses.append(loss_G.item())
                D_losses.append(loss_D.item())
        
        # 각 에폭마다 샘플 이미지 생성
        if (epoch + 1) % 5 == 0 or epoch == 0:
            with torch.no_grad():
                # 고정된 노이즈로 4가지 나이대 샘플 생성
                samples = []
                for age in sample_ages:
                    # 각 나이대별로 16개 이미지 생성
                    age_expanded = age.repeat(16, 1)
                    sample_noise = torch.randn(16, latent_dim, device=device)
                    fake = generator(sample_noise, age_expanded)
                    samples.append(fake)
                
                # 샘플 시각화
                fig, axs = plt.subplots(4, 16, figsize=(32, 8))
                for i, sample in enumerate(samples):
                    for j in range(16):
                        img = sample[j].cpu().detach()
                        img = (img + 1) / 2  # [-1, 1] -> [0, 1]
                        img = img.permute(1, 2, 0)
                        axs[i, j].imshow(img)
                        axs[i, j].axis('off')
                        age_value = sample_ages[i].item() * (dataset.max_age - dataset.min_age) + dataset.min_age
                        if j == 0:
                            axs[i, j].set_title(f"Age: {int(age_value)}")
                
                plt.tight_layout()
                plt.savefig(f"samples_epoch_{epoch+1}.png")
                plt.close()
    
    # 모델 저장
    torch.save(generator.state_dict(), 'age_cgan_generator.pth')
    torch.save(discriminator.state_dict(), 'age_cgan_discriminator.pth')
    
    # 손실 그래프 그리기
    plt.figure(figsize=(10, 5))
    plt.plot(G_losses, label='Generator')
    plt.plot(D_losses, label='Discriminator')
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig('loss_plot.png')
    plt.close()
    
    return generator, discriminator

# 학습 실행
generator, discriminator = train_gan(epochs=200)

# 특정 나이에 대한 이미지 생성 함수
def generate_faces_by_age(generator, ages, n_samples=5):
    """
    ages: 실제 나이 목록 (정규화 전)
    """
    latent_dim = generator.latent_dim
    generator.eval()
    
    fig, axs = plt.subplots(len(ages), n_samples, figsize=(n_samples*2, len(ages)*2))
    
    with torch.no_grad():
        for i, age in enumerate(ages):
            # 나이 정규화
            normalized_age = (age - dataset.min_age) / (dataset.max_age - dataset.min_age)
            age_tensor = torch.tensor([[normalized_age]]).repeat(n_samples, 1).to(device)
            
            # 랜덤 노이즈 생성
            noise = torch.randn(n_samples, latent_dim, device=device)
            
            # 이미지 생성
            fake_images = generator(noise, age_tensor)
            
            # 시각화
            for j in range(n_samples):
                img = fake_images[j].cpu().detach()
                img = (img + 1) / 2  # [-1, 1] -> [0, 1]
                img = img.permute(1, 2, 0)
                
                if len(ages) > 1:
                    axs[i, j].imshow(img)
                    axs[i, j].axis('off')
                    if j == 0:
                        axs[i, j].set_title(f"Age: {age}")
                else:
                    axs[j].imshow(img)
                    axs[j].axis('off')
                    if j == 0:
                        axs[j].set_title(f"Age: {age}")
    
    plt.tight_layout()
    plt.savefig(f"generated_faces.png")
    plt.show()

# 다양한 나이의 얼굴 생성해보기
target_ages = [5, 20, 40, 60, 80]
generate_faces_by_age(generator, target_ages, n_samples=6)
```
